```python
X_train_bert = pd.read_csv('/kaggle/input/x-bert/X_train_bert.csv')
X_test_bert = pd.read_csv('/kaggle/input/x-bert/X_test_bert.csv')
```

```python
X_train_bert = X_train_bert.drop('Unnamed: 0', axis=1)
X_test_bert = X_test_bert.drop('Unnamed: 0', axis = 1)
```

```python
y_train = pd.read_csv('/kaggle/input/y-label/y_train_bert.csv')
y_train = y_train['label']
y_test = pd.read_csv('/kaggle/input/y-label/y_test_bert.csv')
y_test = y_test['label']
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=1000)
model.fit(X_train_bert, y_train)
```

```python
y_pred = model.predict(X_test_bert)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```
